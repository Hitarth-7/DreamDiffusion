```python
import zipfile
import os

zip_path = "MuseData.zip"
extract_dir = "extracted_eeg"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)
```

```python
import glob
import pandas as pd

txt_files = glob.glob(os.path.join(extract_dir, '*.txt'))

txt_file = txt_files[0]

df = pd.read_csv(txt_file, delim_whitespace=True, header=None)

df.columns = ['id','event','device','channel','code','size','data']

df
```

```python
!pip install mne
import mne
df.groupby('channel')
df
```

```python
grouped = df.groupby('channel')
fp1_data = grouped.get_group('FP1')
fp1_data
```

```python
median_size = fp1_data['size'].median()
print(median_size)
```

```python
import numpy as np
fp1_data['data'] = fp1_data['data'].apply(lambda x: np.array(list(map(float, x.split(',')))))
fp1_data
```

```python
!pip install scipy
from scipy.interpolate import interp1d
def interpolate_to_median(signal, target_len):
    original_len = len(signal)
    if original_len == target_len:
        return signal
    x_old = np.linspace(0, 1, original_len)
    x_new = np.linspace(0, 1, target_len)
    interpolator = interp1d(x_old, signal, kind='linear')
    return interpolator(x_new)
```

```python
median_size=int(median_size)
fp1_data['data_interpolated'] = fp1_data['data'].apply(lambda x: interpolate_to_median(x, median_size))
fp1_data
```

```python
data_with_event =fp1_data.apply(
    lambda row: np.insert(row['data_interpolated'], 0, row['event']), axis=1
)


fp1_array = np.vstack(data_with_event.values)
print(fp1_array)
```

```python
fp2_data = grouped.get_group('FP2')
median_size = fp2_data['size'].median()
print(median_size)
fp2_data['data'] = fp2_data['data'].apply(lambda x: np.array(list(map(float, x.split(',')))))
median_size=int(median_size)
fp2_data['data_interpolated'] = fp2_data['data'].apply(lambda x: interpolate_to_median(x, median_size))
data_with_event = fp2_data.apply(
    lambda row: np.insert(row['data_interpolated'], 0, row['event']), axis=1
)


fp2_array = np.vstack(data_with_event.values)
print(fp2_array)
```
