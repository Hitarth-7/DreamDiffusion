Data processing for MU Data

```python
import mne
import matplotlib 
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d
```

```python
df=pd.read_csv('./MU.txt',delimiter='\t',header=None)
df
```

```python
df.columns =['id','event','device','channel','code','size','data']
df
```

```python
df.groupby('channel')
df
```

```python
grouped = df.groupby('channel')
tp9_data = grouped.get_group('TP9')
tp9_data
```

```python
median_size = tp9_data['size'].median()
print(median_size)
```

```python
tp9_data['data'] = tp9_data['data'].apply(lambda x: np.array(list(map(float, x.split(',')))))
tp9_data
```

```python
def interpolate_to_median(signal, target_len):
    original_len = len(signal)
    if original_len == target_len:
        return signal  # No need to interpolate
    x_old = np.linspace(0, 1, original_len)
    x_new = np.linspace(0, 1, target_len)
    interpolator = interp1d(x_old, signal, kind='linear')
    return interpolator(x_new)
```

```python
median_size=int(median_size)
tp9_data['data_interpolated'] = tp9_data['data'].apply(lambda x: interpolate_to_median(x, median_size))
tp9_data
```

```python
data_with_event = tp9_data.apply(
    lambda row: np.insert(row['data_interpolated'], 0, row['event']), axis=1
)


tp9_array = np.vstack(data_with_event.values)
print(tp9_array)
```
