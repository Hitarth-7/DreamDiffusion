```python
eeg_data = du.iloc[:, 6:-1].apply(pd.to_numeric, errors='coerce').values
trigger = du.iloc[:, -1].values

def bandpass_filter(data, lowcut, highcut, fs, order=4):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, data, axis=0)

fs = 250
filtered = bandpass_filter(eeg_data, 0.1, 30.0, fs)

event_indices = [i for i, val in enumerate(trigger) if val != 0]
print("Found", len(event_indices), "stimulus events.")

pre = int(0.2 * fs)
post = int(0.8 * fs)

epochs = []
for idx in event_indices:
    if idx - pre >= 0 and idx + post < len(filtered):
        epoch = filtered[idx - pre : idx + post]
        epochs.append(epoch)

epochs = np.array(epochs)
print("Epochs shape before resampling:", epochs.shape)

target_len = int(np.median([ep.shape[0] for ep in epochs]))

def resample_epoch(epoch, target_len):
    x_old = np.linspace(0, 1, epoch.shape[0])
    x_new = np.linspace(0, 1, target_len)
    if epoch.ndim == 1:
        epoch = epoch[:, np.newaxis]
    return np.array([interp1d(x_old, ch, bounds_error=False, fill_value="extrapolate")(x_new) for ch in epoch.T]).T

resampled = [resample_epoch(e, target_len) for e in epochs]
resampled = [e for e in resampled if e.ndim == 2]
print("Resampled epochs:", len(resampled))

def interpolate_nans(epoch):
    for ch in range(epoch.shape[1]):
        x = np.arange(epoch.shape[0])
        y = epoch[:, ch]
        mask = np.isnan(y)
        if np.all(mask):
            continue
        y[mask] = np.interp(x[mask], x[~mask], y[~mask])
        epoch[:, ch] = y
    return epoch

cleaned = np.array([interpolate_nans(e.copy()) for e in resampled])
print("Cleaned epochs:", cleaned.shape)

def normalize_epoch(epoch):
    scaler = MinMaxScaler()
    return scaler.fit_transform(epoch)

normalized = np.array([normalize_epoch(e) for e in cleaned])
print("Final shape (epochs, time, channels):", normalized.shape)
```
